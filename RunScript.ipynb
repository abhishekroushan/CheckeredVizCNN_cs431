# RSM: Similarity Representations Different layers

```python
model_all_keys= model.state_dict().keys()
print(model_all_keys)
```

```python
def correlation_coefficient(T1, T2):
    numerator = np.mean((T1 - T1.mean()) * (T2 - T2.mean()))
    denominator = T1.std() * T2.std()
    if denominator == 0:
        return 0
    else:
        result = numerator / denominator
        return result
```

```python
conv_weight_keys= ['init_conv.conv.weight',  'conv1.conv.weight', 'conv2.conv.weight',  'conv3.conv.weight',  'conv4.conv.weight']
for weight_key in conv_weight_keys:
    weight= model.state_dict()[weight_key].cpu().numpy()
    nfilters= weight.shape[0]
    corr_mat= np.zeros((nfilters, nfilters))
    for i in range(nfilters):
        m1=weight[i,0,0,:,:]
        for j in range(nfilters):
            m2= weight[j,0,0,:,:]
            corr_mat[i,j]=correlation_coefficient(m1,m2)
    print("--------------Correlation across filters------------------")
    print("Layer: ", weight_key)
    plt.imshow(corr_mat);
    plt.colorbar()
    plt.show()
```

```python
#Experiment Init_conv
from scipy.stats import pearsonr
w= model.state_dict()['fc.weight'].cpu().numpy()
n=w.shape[0]
c= np.zeros((n,n))
for i in range(n):
    for j in range(n):
        corr, p_value = pearsonr(w[i,:], w[j,:])
        c[i,j]= corr
plt.imshow(corr_mat);
plt.colorbar()
plt.show()
```
